```python
import cv2
import numpy as np

def create_image(width, height, rgb_color=(0, 0, 0)):
    image = np.zeros((height, width, 3), np.uint8) 
    color = tuple(reversed(rgb_color))
    image[:] = color
    return image
width, height = 50, 50
red = (255, 0, 0)
green = (0, 255, 0)
blue = (0, 0, 255)
white = (255, 255, 255)
image1 = create_image(width, height, rgb_color=red)
image2 = create_image(width, height, rgb_color=green)
image3 = create_image(width, height, rgb_color=blue)
image4 = create_image(width, height, rgb_color=white)
row1 = np.hstack((image1 , image2))
row2 = np.hstack((image3 , image4))
All = np.vstack((row1 , row2))
All = cv2.resize(All,(200,200))
cv2.imwrite("filename.png",image1)
cv2.imshow("COLOURS",All)

cv2.waitKey(0)
```

```python
import cv2
import numpy as np

n = 0
v = 0
capture = cv2.VideoCapture("Videos/CarController.mp4")
frame_width = int(capture.get(3))
frame_height = int(capture.get(4))
size = (frame_width, frame_height)

def original() :
    saving = False
    while True :
        isTrue, frame = capture.read()  
        cv2.imshow("Video",frame)
        if cv2.waitKey(1)== ord('q')  :
            capture.release()
            cv2.destroyAllWindows()
            break 
        if cv2.waitKey(1) == ord('c') :
            global n
            n=n+1
            cv2.imwrite("frame "+str(n)+".png",frame) 
        if cv2.waitKey(1)== ord('g')  : 
            cv2.destroyAllWindows()
            grey()
        if cv2.waitKey(1)== ord('r')  :
            cv2.destroyAllWindows()
            rotate()
        if cv2.waitKey(1)== ord('h')  :
            cv2.destroyAllWindows()
            HSV()
        if cv2.waitKey(1)== ord('x')  :
            cv2.destroyAllWindows()
            All()
        if cv2.waitKey(1)== ord('s')  :
            global v
            v+=1
            saved = cv2.VideoWriter('clip '+str(v)+'.avi', cv2.VideoWriter_fourcc(*'MJPG'),10, size)
            saving = True
        if saving :
            saved.write(frame)
def grey() :
    saving = False
    while True :
        isTrue, frame = capture.read()
        grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow("Video gray",grey)
        if cv2.waitKey(1)== ord('q')  :
            capture.release()
            cv2.destroyAllWindows()
            break
        if cv2.waitKey(1)== ord('z')  : 
            cv2.destroyAllWindows()
            original()
        if cv2.waitKey(1)== ord('r')  :
            cv2.destroyAllWindows()
            rotate()
        if cv2.waitKey(1) == ord('c') :
            global n
            n+=1
            cv2.imwrite("frame grey " + str(n)+".png",grey) 
        if cv2.waitKey(1)== ord('h')  :
            cv2.destroyAllWindows()
            HSV()
        if cv2.waitKey(1)== ord('x')  :
            cv2.destroyAllWindows()
            All()
        if cv2.waitKey(1)== ord('s')  :
            global v
            v+=1
            saved = cv2.VideoWriter('clip '+str(v)+'.avi', cv2.VideoWriter_fourcc(*'MJPG'),10, size)
            saving = True
        if saving :
            saved.write(grey)
def rotate() :
    saving = False
    while True :
        isTrue, frame = capture.read()
        rot = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
        cv2.imshow("Video rotated",rot)
        if cv2.waitKey(1)== ord('q')  :
            capture.release()
            cv2.destroyAllWindows()
            break
        if cv2.waitKey(1)== ord('z')  : 
            cv2.destroyAllWindows()
            original()
        if cv2.waitKey(1)== ord('g')  :
            cv2.destroyAllWindows() 
            grey()
        if cv2.waitKey(1) == ord('c') :
            global n
            n=n+1
            cv2.imwrite("frame rotated " + str(n)+".png",rot) 
        if cv2.waitKey(1)== ord('h')  :
            cv2.destroyAllWindows()
            HSV()
        if cv2.waitKey(1)== ord('x')  :
            cv2.destroyAllWindows()
            All()
        if cv2.waitKey(1)== ord('s')  :
            global v
            v+=1
            saved = cv2.VideoWriter('clip '+str(v)+'.avi', cv2.VideoWriter_fourcc(*'MJPG'),10, size)
            saving = True
        if saving :
            saved.write(rot)
def HSV() :
    saving = False
    while True :
        isTrue, frame = capture.read()
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        cv2.imshow("Video HSV",hsv)
        saved.write(hsv)
        if cv2.waitKey(1)== ord('q')  :
            capture.release()
            cv2.destroyAllWindows()
            break
        if cv2.waitKey(1)== ord('z')  : 
            cv2.destroyAllWindows()
            original()
        if cv2.waitKey(1)== ord('g')  :
            cv2.destroyAllWindows() 
            grey()
        if cv2.waitKey(1)== ord('r')  :
            cv2.destroyAllWindows()
            rotate()
        if cv2.waitKey(1) == ord('c') :
            global n
            n+=1
            cv2.imwrite("frame HSV " + str(n)+".png",hsv)
        if cv2.waitKey(1)== ord('x')  :
            cv2.destroyAllWindows()
            All() 
        if cv2.waitKey(1)== ord('s')  :
            global v
            v+=1
            saved = cv2.VideoWriter('clip '+str(v)+'.avi', cv2.VideoWriter_fourcc(*'MJPG'),10, size)
            saving = True
        if saving :
            saved.write(hsv)
def All():
    while True :
        isTrue, frame = capture.read()
        cv2.imshow("Video",frame)
        grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow("Video gray",grey)
        rot = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
        cv2.imshow("Video rotated",rot)
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        cv2.imshow("Video HSV",hsv)   
        if cv2.waitKey(1)== ord('q')  :
            capture.release()
            cv2.destroyAllWindows()
            break 
        if cv2.waitKey(1)== ord('g')  : 
            cv2.destroyAllWindows()
            grey()
        if cv2.waitKey(1)== ord('r')  :
            cv2.destroyAllWindows()
            rotate()
        if cv2.waitKey(1)== ord('h')  :
            cv2.destroyAllWindows()
            HSV() 
        if cv2.waitKey(1)== ord('z')  : 
            cv2.destroyAllWindows()
            original()
             
while True :
    original()
    break
capture.release()
cv2.destroyAllWindows() 
cv2.waitKey(0)

   
```

```python
from pickletools import uint8
import numpy as np
import cv2 

img = np.zeros((500,500,3),np.uint8)
img[:] = 0,0,0
clone = img.copy()
circle = []
count = 0 
def mousePoints(event,x,y,flag,params):
    global img,clone,count,circle
    if event == cv2.EVENT_LBUTTONDOWN :
      cv2.circle(img, (x,y), 30, (0,0,255), -1)
      print(x,y)
      count +=1
      circle.append((x,y))
    elif event == cv2.EVENT_RBUTTONDOWN :
      img = clone.copy()
      count -=1
      circle.remove(circle[count-1])
      for i in range (0, count):
        cv2.circle(img, circle[i], 30, (0,0,255), -1)
cv2.namedWindow('black')
while(True):
    cv2.setMouseCallback('black', mousePoints)
    cv2.imshow('black', img)
    k=cv2.waitKey(1) & 0xFF
    if k==27: 
        break
    cv2.imshow('black', img)
cv2.waitKey(0)
```
